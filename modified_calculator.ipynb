```python
import random
import time

def sum_numbers(liste): 
    print("sum of numbers")
    return sum(liste)

def show_numbers(liste):
    print("numbers generated: ")
    return liste

def largest_smallest(liste):
    print("large to small: ")
    return sorted(liste, reverse=True)
    
def smallest_largest(liste):
    print("small to large")
    return sorted(liste)
    
def average_numbers(liste):
    total = sum(liste)
    length = len(liste)
    average = total/length
    return average


while True: 
    
    number = int(input("how many numbers do you want generate: "))
    last_limit = int(input("what is the last limit: "))
    
    
    number_list = [random.randint(0, last_limit) for _ in range(number)]

    print("""services \n[1] show the numbers \n[2] add up the numbers \n[3] sort from smallest to largest \n[4] sort from largest to smallest \n[5] find the mean of the numbers \n[0] quit the app""")

    selected_num = int(input("what action do you want to do: "))
    
    if selected_num == 1: 
        total = show_numbers(number_list)
        
    elif selected_num == 2: 
        total = sum_numbers(number_list)
    
    elif selected_num == 3: 
        print(show_numbers(number_list))
        total = smallest_largest(number_list)
    
    elif selected_num == 4: 
        print(show_numbers(number_list))
        show_numbers(number_list)
        total = largest_smallest(number_list)
        
    elif selected_num == 5: 
        print(show_numbers(number_list))
        total = average_numbers(number_list)
        
    elif selected_num == 0:
        print("exiting the program..")
        time.sleep(2)
        print("exited the program...")
        break
        
    print(total)
        
        
    
    
```
